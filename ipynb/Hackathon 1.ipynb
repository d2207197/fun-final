## Database 功能需求

1. 依據 *任何單詞* 在 *任何位置* 組合搜尋。例如 `'play _ _ role'` 需要搜尋 play 在第一個位置，且，role 在第四個位置。
2. 搜尋結果依據 count 排序

## 已知 HBase 限制

- HBase 僅有單一 indexed 欄位，即 row key。



## 關鍵 HBase 特性

- HBase Table 內的所有資料皆以 row key 內容排序(byte order)。
- HBase 所提供之 scan 可掃描出某個範圍內的 rowkeys。
- HBase 中執行 scan 或 get 皆可指定 column 值來篩選資料。

### Schema 
| ROW KEY                  | COLUMN FAMILY **sel**                     |
| ------------------------ |:------------------------------------------|
| **selected words**#**count**#**hashcode**               | **sel:size-selector** = **"ngram  count"** |




### Schema 
| ROW KEY                  | COLUMN FAMILY **sel**                     |
| ------------------------ |:------------------------------------------|
| **selected words**#**count**#**hashcode**               | **sel:size-selector** = **"ngram  count"** |


## 資料輸入範例

例如如下的 ngram 與 count

`"by love but separated by        67"`

由於共有五個字，selector 為 [0, 1, 2, 3, 4] 的所有子集合共 31 組（不含空集合）。於是會產生 31 個 row。

| ROW KEY                  | COLUMN FAMILY **sel**                     |
| ------------------------ |:------------------------------------------|
| by#67#fe               | sel:5-0  = "by love but separated by  67" |
| love#67#fe               | sel:5-1  = "by love but separated by  67" |
| but#67#fe               | sel:5-2  = "by love but separated by  67" |
| ... | ... |
| but separated#67#fe    | sel:5-23 = "by love but separated by  67" |
| by separated#67#fe     | sel:5-03 = "by love but separated by  67" |
| ... | ... |


## Query 語法的對應

| ROW KEY                  | COLUMN FAMILY **sel**                     |
| ------------------------ |:------------------------------------------|
| by separated#67#fe     | sel:5-03 = "by love but separated by  67" |


於是利用 HBase 的 Scan 功能

`> scan 'tableName', {COLUMNS => ["sel:5-03"], STARTROW => "by separated", STOPRAW => "by separatee"}`

這筆資料正巧可對應到此 query: `"by _ _ separated _"` 




### Schema 
| ROW KEY                  | COLUMN FAMILY **sel**                     |
| ------------------------ |:------------------------------------------|
| **selected words**#**count**#**hashcode**               | **sel:size-selector** = **"ngram  count"** |

## ROW KEY 與 column 設計說明
- Row key
    + **selected words**: 對應到 selector 位置的 words，用於 scan 時作為範圍對應。(4 to 20 Bytse)
    + **count**: 置放在 [selected words] 後方，可使 scan 結果依照 count 排序。(4 Bytes)
    + **hashcode**: 用於避免 [selected words][count] 重複出現，而覆蓋掉另一筆資料。(4 Bytes)
- Column
    + **size-selector**: 5-03 代表資料長度 5，選擇第 0 與 3 個 words。
    
Row key 的所有欄位皆會轉換成 Byte Array。selected words 會依照預先建立的 map 轉換為數字。


# Hadoop Streaming

- HBase 資料預處理與輸入程式由 scala 撰寫，但沒有使用 Hadoop MapReduce API。
- 為了加速資料處理，利用 Hadoop 內建的 Streaming jar 程式，可將任何可執行檔作為 mapper, combiner, reducer。

```
$ yarn jar $HADOOP_MAPRED_HOME/hadoop-streaming.jar -help                                                                              
Usage: $HADOOP_PREFIX/bin/hadoop jar hadoop-streaming.jar [options]
Options:
  -input          <path> DFS input file(s) for the Map step.
  -output         <path> DFS output directory for the Reduce step.
  -mapper         <cmd|JavaClassName> Optional. Command to be run as mapper.
  -combiner       <cmd|JavaClassName> Optional. Command to be run as combiner.
  -reducer        <cmd|JavaClassName> Optional. Command to be run as reducer.
  -file           <file> Optional. File/dir to be shipped in the Job jar file.
                  Deprecated. Use generic option "-files" instead.
  -inputformat    <TextInputFormat(default)|SequenceFileAsTextInputFormat|JavaClassName>
                  Optional. The input format class.
  -outputformat   <TextOutputFormat(default)|JavaClassName>
                  Optional. The output format class.
  -partitioner    <JavaClassName>  Optional. The partitioner class.
   -numReduceTasks <num> Optional. Number of reduce tasks.
  -inputreader    <spec> Optional. Input recordreader spec.
...
```


# Hadoop Streaming 執行範例

```
$ yarn jar $HADOOP_MAPRED_HOME/hadoop-streaming.jar  \
   -files web1t_unigrams_300000up.json,target/scala-2.10/default-assembly-0.1-SNAPSHOT.jar,hbase-site.xml  \
   -input /corpus/Web1T/2gms,/corpus/Web1T/3gms,/corpus/Web1T/4gms,/corpus/Web1T/5gms \
   -output putNgram-output \
   -mapper 'java -jar hbasePutNgram-assembly-0.1-SNAPSHOT.jar web1t_unigrams_300000up.json web1t-linggle'
```

- 利用 `-files` 參數將程式執行過程所需的額外檔案上傳。
- `-input` 指定輸入資料
- `-output` 指定輸出資料，由於資料直接到 HBase，此 output 僅用來觀察資料
- `-mapper` 指定 Mapper 指令


# Hadoop Streaming 優缺點

## 優
- 可使用任何語言撰寫 Mapper, Reducer, Combiner。也可使用現成的執行檔，例如 wc, cat, uniq, grep, sort。
- 可輕鬆 local 測試
- debug 方便

## 缺
- 若不使用 Hadoop MapReduce API 撰寫 Mapper, 將受限在 `<Text, Text>` 資料格式
- 無法自定 GroupComparator 與 SortComparator。
- 多 Job 需獨立跑多個 streaming.jar
